```python
gender_df = pd.read_csv('gender.csv', header=0)
gender_df.fillna(0, inplace=True)
gender_df.head()
```

```python
def proportion_calculator(permits, summits):
    return permits / summits * 100
```

```python
import re 

def get_height_from_name(peak_name):
    pat = r'.*?\((.*)m.*'
    match = re.search(pat, peak_name)
#     print(f"match: {match.group(1)}")
    return (int)(match.group(1))   

def filter_on_height(df, height):
    df = df[df.apply(lambda x: get_height_from_name(x['Peak']) > height, axis=1)]
    return df
```

```python
def calculate_average(list_of_permits, list_of_summits):
    print(list_of_permits)
    print(list_of_summits)
    list_of_proportions = []
    for (i,j) in zip(list_of_permits, list_of_summits):
        try:
            list_of_proportions.append(j/i)
        except ZeroDivisionError:
            list_of_proportions.append(0)
    return sum(list_of_proportions)/len(list_of_proportions) * 100
```

```python
def get_proportion_based_on_height(df, height):
    new_df = filter_on_height(gender_df, height)
    prop = calculate_average(new_df['Female'].tolist(), new_df['Female.1'].tolist())
    print(f"Proportion of climbers who successfully summitted: {prop}")
```

```python
get_proportion_based_on_height(gender_df, 8500)
```
